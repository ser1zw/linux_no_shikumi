```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def show_graphs(filename, title):
    data = pd.read_table(filename, header=3)
    LABEL_PROCESS_ID, LABEL_ELAPSED_TIME, LABEL_PROGRESS = data.columns
    process_ids = np.sort(data[LABEL_PROCESS_ID].unique())

    plt.figure(figsize=[20, 5])

    plt.subplot(1, 2, 1)
    for p in process_ids:
        prot_data = data[data[LABEL_PROCESS_ID] == p]
        x = prot_data[LABEL_ELAPSED_TIME]
        y = prot_data[LABEL_PROCESS_ID]
        plt.scatter(x, y, s=5)

    plt.xlim(0, np.max(data[LABEL_ELAPSED_TIME]))
    plt.ylim(0, 4)
    plt.title('Elapsed time - Process#')
    plt.xlabel(LABEL_ELAPSED_TIME)
    plt.ylabel(LABEL_PROCESS_ID)
    plt.legend(process_ids)

    plt.subplot(1, 2, 2)
    for p in process_ids:
        prot_data = data[data[LABEL_PROCESS_ID] == p]
        x = prot_data[LABEL_ELAPSED_TIME]
        y = prot_data[LABEL_PROGRESS]
        plt.scatter(x, y, s=1)

    plt.xlim(0, np.max(data[LABEL_ELAPSED_TIME]))
    plt.ylim(0, 100)
    plt.title('Elapsed time - Progress')
    plt.xlabel(LABEL_ELAPSED_TIME)
    plt.ylabel(LABEL_PROGRESS)
    plt.legend(process_ids)

    plt.suptitle(title)
    plt.subplots_adjust(wspace=0.2)

    plt.show()
```

```python
# taskset -c 0 ./sched 1 100 1 > exp4A.tsv
show_graphs('exp4A.tsv', 'Exp 4-A: Core=1, n=1, total=100, resoltion=1')
```
