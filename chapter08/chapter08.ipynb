```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
import pandas as pd

LABLE_IO_SIZE = 'log2(I/O size[KB])'
LABLE_THROUGHPUT = 'Throughput [MB/s]'

def elapse2throughput(s):
    minute, second = map(float, s.removesuffix('s').split('m'))
    return 64.0 / (minute * 60 + second)

def add_graph(filename, title):
    df = pd.read_table(filename, sep = ' ', header = None, names = (LABLE_IO_SIZE, LABLE_THROUGHPUT))

    df[LABLE_IO_SIZE] = df[LABLE_IO_SIZE].map(math.log2)
    df[LABLE_THROUGHPUT] = df[LABLE_THROUGHPUT].map(elapse2throughput)

    plt.scatter(df[LABLE_IO_SIZE], df[LABLE_THROUGHPUT])
    plt.title(title)
    plt.xlabel(LABLE_IO_SIZE)
    plt.ylabel(LABLE_THROUGHPUT)
    return plt

def show_graph(filename, title):
    add_graph(filename, title)
    plt.show()

def show_graphs(filename1, legend1, filename2, legend2, title):
    add_graph(filename1, title)
    add_graph(filename2, title)
    plt.legend([legend1, legend2])
    plt.show()
```

```python
show_graph('io-result/log-hdd-seq-read.txt', 'HDD sequential read (No acceleration)')
show_graph('io-result/log-hdd-seq-write.txt', 'HDD sequential write (No acceleration)')
```

```python
show_graphs('io-result/log-hdd-seq-read.txt', 'sequential', 'io-result/log-hdd-rand-read.txt', 'random', 'HDD read (No acceleration)')
show_graphs('io-result/log-hdd-seq-write.txt', 'sequential', 'io-result/log-hdd-rand-write.txt', 'random', 'HDD write (No acceleration)')
```
